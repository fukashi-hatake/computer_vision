## EasyOCR  

- OCR = Optical Character Recognition

> https://github.com/JaidedAI/EasyOCR

- Blog: https://blog.roboflow.com/ocr-on-videos/ 

```python
import easyocr 

from PIL import Image
```

```python
reader = easyocr.Reader(['en']) 
```

```python
image_link = "twitter-cover-3.png"

Image.open(image_link)
```

```python
reader.readtext(image_link)
```
